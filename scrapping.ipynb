```python
#Load needed modules
from bs4 import BeautifulSoup
import requests
import pandas as pd

#Link to the site where we want to scrape
site = f"https://www.realmadrid.com/en/football/squad"

r = requests.get(site)
page = r.text

players_menu = BeautifulSoup(page, 'html.parser')

#Creating an empty list
players_list = []
if r.status_code == 200:
    
    #selecting the player div
    players = players_menu.select(".m_player_info_content")
    
    for p in players[0:25]:
    
        #Get the name of the player
        player_name = p.select("span")[0].get_text().strip()

        #Get the link to the profile of the player
        player_link = p.find("a").attrs['href']

        #Complate the url with player_link
        url = f"https://www.realmadrid.com{player_link}"

        #Create a var with dict
        new_player = {"player_name" : player_name, "url" : url}

        #add our data to the list
        players_list.append(new_player)
```

```python

for count, url in enumerate(players_list):
    r_p = requests.get(url["url"])
    page_p = r_p.text
    player_html = BeautifulSoup(page_p, 'html.parser')
    
    #Checking if we can connect to the website
    if r_p.status_code == 200:
        
        #Selecting the place of birth of the player
        place_of_birth = player_html.select(".data_info span")[1].get_text().strip()
        
        #Selecting the birthday of the player
        birthday = player_html.select(".data_info span")[2].get_text().strip()
        
        #Selecting the weight of the player
        weight = player_html.select(".col-1-3 .data_info span")[0].get_text().replace("kg."," ").strip()
        
        #Updating the list with our new data
        players_list[count].update({"birthday" : birthday, "place_of_birth" : place_of_birth, "weight" : weight})
    else:
        #Printing error when we cant find the url
        print("We can not connect to the website!")
```

```python
#Dataframe and resorting the columns
df = pd.DataFrame(players_list, columns = ['player_name', 'url', 'birthday', 'place_of_birth', 'weight'])
```
